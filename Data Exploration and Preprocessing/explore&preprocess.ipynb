Exploration of the file can be done using the loc and iloc functions 
Loc allows use of string index and integer index to identify the specific rows

Whereas the iloc uses only integers to identify the specific rows


```python
#using the iloc function
data_csv.iloc[4]     #4 represents the fourth row
```

```python
#using iloc with both the row and column
data_csv.iloc[4,3]     #outputs the value in the 4th row in column 3(total_rooms)
```

```python
#using the loc function 
data_csv.loc[0]
```

```python
#using the loc function with both rows and columns
data_csv.loc[0,"latitude"]
```
